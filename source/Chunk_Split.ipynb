```
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.decomposition import TruncatedSVD
from sklearn.feature_selection import SelectKBest, chi2, mutual_info_classif
from sklearn.feature_extraction.text import TfidfVectorizer, HashingVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import Perceptron
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.preprocessing import MaxAbsScaler
from sklearn import metrics
from tempfile import TemporaryFile

data_train = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Train_RUSMOTE.csv", index_col=None)
data_test = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Validate_RUSMOTE.csv", index_col=None)

print(data_train.dtypes)
sc = MinMaxScaler()
data_train['Coord_Conjunctions'] = sc.fit_transform(data_train['Coord_Conjunctions'].values.reshape(-1, 1))
data_train['Characters'] = sc.fit_transform(data_train['Characters'].values.reshape(-1, 1))
data_train['Quotes'] = sc.fit_transform(data_train['Quotes'].values.reshape(-1, 1))
data_train['Paragraphs'] = sc.fit_transform(data_train['Paragraphs'].values.reshape(-1, 1))
data_train['Semi_Colons'] = sc.fit_transform(data_train['Semi_Colons'].values.reshape(-1, 1))
data_train['Prepositions'] = sc.fit_transform(data_train['Prepositions'].values.reshape(-1, 1))
data_train['Possesive_Pronoun'] = sc.fit_transform(data_train['Possesive_Pronoun'].values.reshape(-1, 1))
data_train['Personal_Pronoun'] = sc.fit_transform(data_train['Personal_Pronoun'].values.reshape(-1, 1))
data_train['Periods'] = sc.fit_transform(data_train['Periods'].values.reshape(-1, 1))
data_train['Hyphens'] = sc.fit_transform(data_train['Hyphens'].values.reshape(-1, 1))
data_train['Commas'] = sc.fit_transform(data_train['Commas'].values.reshape(-1, 1))
data_train['Colons'] = sc.fit_transform(data_train['Colons'].values.reshape(-1, 1))
data_train['Female_Orientation'] = sc.fit_transform(data_train['Female_Orientation'].values.reshape(-1, 1))
data_train['Male_Orientation'] = sc.fit_transform(data_train['Male_Orientation'].values.reshape(-1, 1))
data_train['Average_Sentence_Length'] = sc.fit_transform(data_train['Average_Sentence_Length'].values.reshape(-1, 1))
data_train['Flesch_Score'] = sc.fit_transform(data_train['Flesch_Score'].values.reshape(-1, 1))
data_train['Interjections'] = sc.fit_transform(data_train['Interjections'].values.reshape(-1, 1))

data_test['Coord_Conjunctions'] = sc.fit_transform(data_test['Coord_Conjunctions'].values.reshape(-1, 1))
data_test['Characters'] = sc.fit_transform(data_test['Characters'].values.reshape(-1, 1))
data_test['Quotes'] = sc.fit_transform(data_test['Quotes'].values.reshape(-1, 1))
data_test['Paragraphs'] = sc.fit_transform(data_test['Paragraphs'].values.reshape(-1, 1))
data_test['Semi_Colons'] = sc.fit_transform(data_test['Semi_Colons'].values.reshape(-1, 1))
data_test['Prepositions'] = sc.fit_transform(data_test['Prepositions'].values.reshape(-1, 1))
data_test['Possesive_Pronoun'] = sc.fit_transform(data_test['Possesive_Pronoun'].values.reshape(-1, 1))
data_test['Personal_Pronoun'] = sc.fit_transform(data_test['Personal_Pronoun'].values.reshape(-1, 1))
data_test['Periods'] = sc.fit_transform(data_test['Periods'].values.reshape(-1, 1))
data_test['Hyphens'] = sc.fit_transform(data_test['Hyphens'].values.reshape(-1, 1))
data_test['Commas'] = sc.fit_transform(data_test['Commas'].values.reshape(-1, 1))
data_test['Colons'] = sc.fit_transform(data_test['Colons'].values.reshape(-1, 1))
data_test['Female_Orientation'] = sc.fit_transform(data_test['Female_Orientation'].values.reshape(-1, 1))
data_test['Male_Orientation'] = sc.fit_transform(data_test['Male_Orientation'].values.reshape(-1, 1))
data_test['Average_Sentence_Length'] = sc.fit_transform(data_test['Average_Sentence_Length'].values.reshape(-1, 1))
data_test['Flesch_Score'] = sc.fit_transform(data_test['Flesch_Score'].values.reshape(-1, 1))
data_test['Interjections'] = sc.fit_transform(data_test['Interjections'].values.reshape(-1, 1))

#data_train.drop('book_id', axis=1, inplace=True)
#data_test.drop('book_id', axis=1, inplace=True)
data_train = data_train[data_train['genre'] != 'Unlabelled']

Y_train = data_train.pop('genre')
Y_test = data_test.pop('genre')
X_train = data_train.values
X_test = data_test.values
print(data_train.head())

#clf = RandomForestClassifier()
clf = MLPClassifier(hidden_layer_sizes= (15,) , max_iter=1000, warm_start=True, verbose=1, learning_rate = 'adaptive')
clf.fit(X_train, Y_train)
pred = clf.predict(X_test)
score = metrics.accuracy_score(Y_test, pred)
score1 = metrics.classification_report(Y_test, pred , zero_division=1)
print(score1)
print("accuracy:   %0.5f" % score)
```
